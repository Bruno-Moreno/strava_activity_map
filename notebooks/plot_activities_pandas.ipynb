# Plot Strava Activities 

```python
import pandas as pd
import os 
import plotly.express as px
```

## Read Data

```python
data_path = "..//data"
```

```python
data_files = [f for f in os.listdir(data_path) if os.path.isfile(os.path.join(data_path, f)) and f.endswith('csv')]
data_files
```

```python
data_files = [f for f in os.listdir(data_path) if os.path.isfile(os.path.join(data_path, f)) and f.endswith('csv')]
df = pd.DataFrame() 
for d in data_files:
    df_ = pd.read_csv(f'{data_path}//{d}')
    df = pd.concat([df, df_], axis = 0)
df
```

## Plot Values

```python
import plotly.express as px
import pandas as pd

# Create a scatter plot on a map
fig = px.scatter_mapbox(
    df, 
    lat='lat', 
    lon='long',
    hover_name='activity_name',  # Assuming you have a column with run IDs
    zoom=10,
    color = 'activity_type'
)
fig.update_layout(mapbox_style="open-street-map")
fig.write_html('..//templates//strava_activities.html')
fig.show()
# Update layout for the map
```
